**Procesamiento de imagen con python**

```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image
```

```python
urlDeImagen = 'https://upload.wikimedia.org/wikipedia/commons/thumb/8/82/Circle-yellow.svg/1200px-Circle-yellow.svg.png'
imagenDeInternet = Image.open(urllib.request.urlopen(urlDeImagen))

circuloAmarillo = np.array(imagenDeInternet)
```

```python
plt.imshow(circuloAmarillo)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(circuloAmarillo)
```

```python
carita = np.array(circuloAmarillo)

plt.figure(figsize=(10,20))
plt.imshow(carita)
```

```python

# pintar el ojo izquierdo
carita[390:600,710:820,0] = 0 #R
carita[390:600,710:820,1] = 0 #G
carita[390:600,710:820,2] = 0 #B

plt.figure(figsize=(10,20))
plt.imshow(carita)
```
